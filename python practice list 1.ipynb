````python
def is_leap_year(year):
    """
    Check if a year is a leap year or not.
    """
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

# Input year from the user
year = int(input("Enter a year: "))

# Check if the year is a leap year or not
if is_leap_year(year):
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
````

````python
def sum_of_natural_numbers(n):
    """
    Calculate the sum of natural numbers up to n.
    """
    # Initialize sum
    total = 0

    # Iterate through numbers from 1 to n
    for i in range(1, n + 1):
        total += i

    return total

# Input the value of n from the user
n = int(input("Enter a positive integer: "))

# Check if the input is valid
if n <= 0:
    print("Please enter a positive integer.")
else:
    # Calculate and print the sum of natural numbers up to n
    sum_natural = sum_of_natural_numbers(n)
    print("The sum of natural numbers up to", n, "is:", sum_natural)
````

````python
def is_prime(num):
    """
    Check if a number is prime.
    """
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def print_prime_in_interval(start, end):
    """
    Print all prime numbers within the given interval [start, end].
    """
    print("Prime numbers between", start, "and", end, "are:")
    for num in range(start, end + 1):
        if is_prime(num):
            print(num, end=" ")

# Input the interval from the user
start = int(input("Enter the start of the interval: "))
end = int(input("Enter the end of the interval: "))

# Check if the input interval is valid
if start >= end:
    print("Invalid interval. Start value should be less than end value.")
else:
    # Print all prime numbers within the interval
    print_prime_in_interval(start, end)
````

````python
def factorial(n):
    """
    Calculate the factorial of a number.
    """
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

# Input the number from the user
num = int(input("Enter a non-negative integer: "))

# Check if the input is valid
if num < 0:
    print("Factorial is not defined for negative numbers.")
elif num == 0:
    print("Factorial of 0 is 1.")
else:
    # Calculate and print the factorial of the number
    fact = factorial(num)
    print("Factorial of", num, "is:", fact)
````

````python
#App1:Design a python application for Cricket Game. Take two teams Computer and SYMCA. Take random number from 30-50 for computer team score. Use random module ranging from 0 to 6 for balling. If number 0 occurs in random number player is out. Game should of 5 overs. If all the players are out, declare SYMCA All Out. Display score player wise. And display final match result between Computer and SYMCA team.
import random

class CricketGame:
    def __init__(self, team1, team2):
        self.team1 = team1
        self.team2 = team2
        self.scores = {team1: [], team2: []}

    def play_game(self):
        print("Cricket Match between {} and {}".format(self.team1, self.team2))
        print("5 overs match begins!")

        for _ in range(5):
            self.play_over(self.team1)
            self.play_over(self.team2)

        self.display_scores()

    def play_over(self, team):
        total_runs = 0
        players_out = 0

        print("\n{} is batting now".format(team))

        for _ in range(6):
            runs = random.randint(0, 6)
            if runs == 0:
                print("Player is out!")
                players_out += 1
            total_runs += runs

        self.scores[team].append(total_runs)
        print("Total runs scored by {} in this over: {}".format(team, total_runs))

        if players_out == 6:
            print("All players are out for {}".format(team))

    def display_scores(self):
        print("\nFinal Scores:")
        for team, runs_list in self.scores.items():
            total_runs = sum(runs_list)
            print("{}: {}".format(team, total_runs))

        winner = max(self.scores, key=lambda k: sum(self.scores[k]))
        print("\n{} wins the match!".format(winner))

# Create an instance of the CricketGame class
cricket_game = CricketGame("Computer", "SYMCA")

# Start the game
cricket_game.play_game()

#This program simulates a cricket match between two teams, "Computer" and "SYMCA". It runs for 5 overs. Each over consists of 6 balls, and the runs scored are generated randomly between 0 and 6 for each ball. If the generated run is 0, the player is considered out. After 5 overs, the total score of each team is displayed, and the team with the highest total score wins the match.
````

````python
#App2:-Design a python application for rolling dice. Take 2 players and give 10 turns to both the player simultaneously. If 4 or 6 arises while rolling a dice give bonus turn to that player. Display player wise score after every turn. Calculate score of both the players and display final score and declare winner.
import random

class DiceRollGame:
    def __init__(self, player1, player2):
        self.player1 = player1
        self.player2 = player2
        self.scores = {player1: 0, player2: 0}

    def play_game(self):
        print("Dice Rolling Game between {} and {}".format(self.player1, self.player2))
        print("Each player gets 10 turns!")

        for _ in range(10):
            self.play_turn(self.player1)
            self.play_turn(self.player2)

        self.display_scores()
        self.declare_winner()

    def play_turn(self, player):
        print("\n{}'s turn:".format(player))
        roll = random.randint(1, 6)
        print("Dice roll:", roll)

        if roll in (4, 6):
            print("Bonus turn for {}!".format(player))
            self.scores[player] += roll
            self.play_turn(player)
        else:
            self.scores[player] += roll

        print("Score after this turn:", self.scores[player])

    def display_scores(self):
        print("\nFinal Scores:")
        for player, score in self.scores.items():
            print("{}: {}".format(player, score))

    def declare_winner(self):
        winner = max(self.scores, key=self.scores.get)
        print("\n{} wins the game with a score of {}!".format(winner, self.scores[winner]))

# Create an instance of the DiceRollGame class
game = DiceRollGame("Player 1", "Player 2")

# Start the game
game.play_game()

#This program simulates a dice rolling game between two players. Each player gets 10 turns. During each turn, a player rolls a dice, and if they get a 4 or a 6, they receive a bonus turn. The scores for each player are accumulated after each turn, and the final scores are displayed at the end. The player with the highest score wins the game.
````
